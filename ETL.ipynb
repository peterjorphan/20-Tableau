```python
import pandas as pd
import os
```

```python
# Get list of all the csv files in the folder
path = './Resources/'
filelist = os.listdir(path) 
filelist = [x for x in filelist if '.csv' in x]
filelist
```

```python
# find the number of total records in the files
n=0
for file in filelist:
    n += len(pd.read_csv(path+file))
n
```

# Sample Selection
Based on the total number of records it was decided to select a random sample as a percentage of them by iterrating based on a specific step, and saving it to a new csv file.

### Estimate the sample size based on the step used

```python
step = 10
```

```python
size = int((n-1)/step)+1
print (f"Population Size: {n:,}")
print (f"Estimated Sample Size: {size:,}")
print (f"Sample % of total: {size/n:.2%}")
```

### Get the column headers off the most recent file

```python
lastfile = pd.read_csv(path+filelist[-1])
columns = lastfile.columns
```

### Merge all files into one dataframe and generate sample

```python
# df = pd.DataFrame()
# for file in filelist:
#     file_df = pd.read_csv('Resources/'+file)
#     file_df.columns = columns
#     df = df.append(file_df, sort=False)
# len(df)
```

```python
# # Create a sample dataframe with the step defined earlier
# sample_df = df.iloc[::step, :]
# len(sample_df)
```

```python
# Merge files applying the step first (runs faster than merging first and then selecting the sample due to less memory used)
sample_df = pd.DataFrame()
m = 0 # needs to be a number < step
f = 0
for file in filelist:
    file_df = pd.read_csv(path+file)
    file_df.columns = columns
    sample_df = sample_df.append(file_df.iloc[m::step, :], sort=False, ignore_index=True)
    m = (len(file_df)-1-m) % step
    f += 1 
    print(f)
len(sample_df)
```

```python
sample_df
```

```python
# Recalculate the sample size to validate accuracy
size_new = len(sample_df)
print (f"Sample Size: {size_new:,}")
print (f"Difference From Estimated Sample Size: {(size_new - size):,}")
print (f"Sample % of total: {size/n:.2%}")
```

```python
# Save sample dataframe to a csv file
sample_df.to_csv('./sample-citibike-tripdata.csv', index=False)
```
